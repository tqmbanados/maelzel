```python
# This should return Bb and A and not A# and A!!

Chord(["4A", "4Bb"], 1)
```

```python
# This should return G#, A and Bb

Chord(["4A", "4Bb", "4G#"], 1)
```

```python
# There is no good solution for this

Chord("4G 4G# 4A 4Bb".split(), 1)
```

```python
# Only solution: F# G# A Bb

Chord("4Gb 4G# 4A 4Bb".split(), 1)
```

```python
# Only solution: F G# A Bb-

Chord("4F 4G# 4A 4A+".split(), 1)
```

```python
# Only solution: F G# A- Bb-

Chord("4F 4G# 4A- 4A+".split(), 1)
```
