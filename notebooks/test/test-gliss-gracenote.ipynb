```python
from maelzel.core import *
setTempo(60)
ch = Chain("""
    C4-:0.5
    4D+,4Eb:0.5
    4E-:0.5
    4F-:0.5
""")
ch2 = ch.cycle(totaldur=20)
for i, ev in enumerate(ch2):
    if i%3 == 0:
        ev.gliss = True
    if i%5 == 0:
        ev.gliss = False
        ev.setPlay(end=ev.dur*0.25, fade=(0.005, 0.05))
        ev.addSymbol('staccato')
    if i%7 == 0:
        ev.dur = ev.dur * F(3, 2)
    if i%11 == 0:
        ev.dur = ev.dur / 2
        
ch2.show()
ch2.play(instr='sin')

# TODO: revise rules to break syncopations 
```
